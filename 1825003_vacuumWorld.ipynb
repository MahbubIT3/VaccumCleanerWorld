Here I implement the Vacuum Cleaner World functionality.


1.   Here is an agent who can perform actions **Suck**, **Move** and **Stay** .
2.   Two Environments **Dirty** and **Clear**.
3.   Two States **State1** and **State2**.

How it will work?
1.   The environment agent's state is random.
2.   According to the rules the agent do some actions in the states.
3. The agent can Suck, Move and Stay in the state.

Here I give two examples:

1. 
```
AGENT: Vacuum Cleaner
ACTIONS:  ['suck', 'move', 'stay']
ENVIRONMENT:  ['dirty', 'clear']
STATE:  ['state1', 'state2']
```

```
The state2  is  clear
```

````python
import random

actions = ['suck', 'move', 'stay']
environment = ['dirty', 'clear']
state = ['state1', 'state2']

print("AGENT: Vacuum Cleaner")
print("ACTIONS: ",actions)
print("ENVIRONMENT: ",environment)
print("STATE: ",state, "\n")
r_environment = random.choice(environment)
r_state = random.choice(state)

# Functions

def moveToState2(r_environment):
      r_state = "state2"
      print("\n")
  
      print("The next", r_state, "is", r_environment)
      actions = "move"
      r_state = "state2"
      print("The agent perform action", actions, "to the state2", r_state)
      suck("state2")

def moveToState1(r_environment):
  r_state = "state1"
  print("\n")
  
  print("The next", r_state, "is ", r_environment)
  actions = "move"
  r_state = "state1"
  print("So, the agent perform action", actions, "to the", r_state)
  suck("state1")

def suck(r_state):
  actions = "suck"
  r_environment = "clear"
  print("The agent perform action", actions, " in ", r_state," and the environment is", r_environment ,"now")
  r_environment = random.choice(environment)


def stayAtState1():
  r_environment = "clear"
  r_state ="state2"
  print("\n")
  
  print("The next", r_state, "is clear", r_environment)
  actions = "stay"
  r_state = "state1"
  print("So, the agent perform action ", actions, "at the", r_state)
  


def stayAtState2():
  r_environment = "clear"
  r_state = "state1"
  print("\n")
  print("The next state", r_state, "is clear", r_environment)
  actions = "stay"
  r_state = "state2"
  print("So, the agent perform action ", actions, "at the ", r_state)
  
def dirtyState1():
    r_state = "state1"
    r_environment = "dirty"
    print("The", r_state, " is ", r_environment)
    suck(r_state)
    r_environment = random.choice(environment)

    if(r_environment == "dirty"):
      r_environment = "dirty"
      r_state = "state2"
      moveToState2("dirty")
    else:
      r_environment = "clear"
      r_state = "state2"

      if(r_environment == "clear"):
        r_state = "state1"
        stayAtState1()
      
      else:
        r_state = "state2"
        moveToState2("dirty")

def dirtyState2():
    r_state = "state2"
    r_environment = "dirty"
    print("The", r_state, " is ", r_environment)
    suck(r_state)

    r_environment = random.choice(environment)

    if(r_environment == "dirty"):
      r_environment = "dirty"
      r_state = "state1"
      moveToState1("dirty")
    else:
      r_environment = "clear"
      r_state = "state1"

      if(r_environment == "clear"):
        r_state = "state2"
        stayAtState2()
      
      else:

        r_state = "state1"
        moveToState1("dirty")


# For state1
if(r_state == "state1"):
  print("The agent is in ",r_state)
  if(r_environment == "dirty"):
    
    dirtyState1()
  else:
    print("The", r_state, " is ", r_environment)
    r_environment = random.choice(environment)

    if(r_environment == "clear"):
      r_state = "state1"
      stayAtState1()
      
    else:

        r_state = "state2"
        moveToState2("dirty")
      

# For State2
else:
  print("\nThe agent is in",r_state)
  if(r_environment == "dirty"):
    dirtyState2()


  else:
      print("The", r_state, " is ", r_environment)
      r_environment = random.choice(environment)

      if(r_environment == "clear"):
        r_state = "state2"
        stayAtState2()
      
      else:
        r_state = "state1"
        moveToState1("dirty")

````
